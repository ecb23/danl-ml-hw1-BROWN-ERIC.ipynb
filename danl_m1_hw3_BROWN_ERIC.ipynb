```python
#QUESTION 6

#Select employees who have â€œOT Hoursâ€ greater than 0,
#calculate their â€œOT Rateâ€ (Total OT Paid / OT Hours),
#and then find the employee with the highest â€œOT Rateâ€.
#Display their full name and OT Rate.
#Hint: Start with filtering observations in nyc_payroll
#with the following lines:



# Boolean Series of whether or not 'OT Hours' is greater than 0.
pos_OT_hours = nyc_payroll['OT Hours'] > 0

# Keeping only the observations with pos_OT_hours == True
with_ot = nyc_payroll[ pos_OT_hours ]


ot_greater_than_zero = nyc_payroll[nyc_payroll['OT Hours'] > 0]

# calculating OT rate


ot_greater_than_zero['OT Rate'] = ot_greater_than_zero['Total OT Paid'] / ot_greater_than_zero['OT Hours']


# highest OT rate: ?

# nyc_payroll.sort_values(["OT rate"] .head(1)

highest_ot_rate_record = ot_greater_than_zero.loc[ot_greater_than_zero['OT Rate'].idxmax()]
nyc_payroll








```
